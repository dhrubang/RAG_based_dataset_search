# EDA + Feature Engineering

```python

# Load dataset
df = pd.read_csv('kaggle_datasets_metadata.csv')

# Basic info
print("Dataset shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nMissing values per column:")
print(df.isnull().sum())

# Clean and preprocess text data
df['about_dataset'] = df['about_dataset'].fillna('').str.lower()
df['dataset_name'] = df['dataset_name'].fillna('').str.lower()

# Initialize tokenizer and stopwords
tokenizer = RegexpTokenizer(r'\w+')
stop_words = set(stopwords.words('english'))

def preprocess(text):
    tokens = tokenizer.tokenize(text)
    return [word for word in tokens if word not in stop_words and len(word) > 3]

df['tokens'] = df['about_dataset'].apply(preprocess)

# Word frequency tags
all_tokens = [token for tokens in df['tokens'] for token in tokens]
tag_counts = Counter(all_tokens)
top_tags = dict(sorted(tag_counts.items(), key=lambda x: x[1], reverse=True)[:15])

# Plot top tags
plt.figure(figsize=(12, 8))
sns.barplot(x=list(top_tags.values()), y=list(top_tags.keys()), palette='viridis')
plt.title('Top 15 Dataset Tags', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Tag', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()

# Word cloud
text = ' '.join(df['about_dataset'])
additional_stopwords = {'data', 'dataset', 'use', 'used', 'including', 'features', 'feature'}

wordcloud = WordCloud(
    width=1000,
    height=600,
    background_color='white',
    stopwords=stop_words.union(additional_stopwords),
    colormap='viridis',
    max_words=100
).generate(text)

plt.figure(figsize=(14, 8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Dataset Descriptions', fontsize=16, pad=20)
plt.show()

# Sentiment analysis
def get_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

df['sentiment'] = df['about_dataset'].apply(get_sentiment)

plt.figure(figsize=(10, 6))
sns.histplot(df['sentiment'], bins=20, kde=True)
plt.title('Sentiment Distribution of Dataset Descriptions', fontsize=16)
plt.xlabel('Sentiment Polarity (-1 to 1)', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.axvline(x=0, color='r', linestyle='--')
plt.show()

# Most common words in dataset names
dataset_names = ' '.join(df['dataset_name'].dropna()).split()
name_word_counts = Counter([word for word in dataset_names if len(word) > 3 and word not in stop_words])

plt.figure(figsize=(12, 8))
sns.barplot(
    x=[count for word, count in name_word_counts.most_common(15)],
    y=[word for word, count in name_word_counts.most_common(15)],
    palette='magma'
)
plt.title('Top 15 Words in Dataset Titles', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Word', fontsize=14)
plt.show()

# URL domain analysis
df['domain'] = df['dataset_url'].apply(lambda x: x.split('/')[2] if pd.notnull(x) else '')
domain_counts = df['domain'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=domain_counts.values, y=domain_counts.index, palette='rocket')
plt.title('Top 10 Domain Sources for Datasets', fontsize=16)
plt.xlabel('Count', fontsize=14)
plt.ylabel('Domain', fontsize=14)
plt.show()

# Description length analysis
df['desc_length'] = df['about_dataset'].apply(lambda x: len(x.split()))
plt.figure(figsize=(10, 6))
sns.histplot(df['desc_length'], bins=30, kde=True)
plt.title('Distribution of Description Lengths (Words)', fontsize=16)
plt.xlabel('Description Length (words)', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()

# Description length vs sentiment
plt.figure(figsize=(10, 6))
sns.scatterplot(x='desc_length', y='sentiment', data=df, alpha=0.6)
plt.title('Description Length vs. Sentiment', fontsize=16)
plt.xlabel('Description Length (words)', fontsize=14)
plt.ylabel('Sentiment Polarity', fontsize=14)
plt.show()

# Extract years mentioned in descriptions
years = re.findall(r'\b(19|20)\d{2}\b', ' '.join(df['about_dataset']))
year_counts = Counter(years)
sorted_years = dict(sorted(year_counts.items(), key=lambda x: x[0]))

plt.figure(figsize=(12, 6))
sns.lineplot(x=list(sorted_years.keys()), y=list(sorted_years.values()), marker='o')
plt.title('Mention of Years in Dataset Descriptions', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

# --- LDA Topic Modeling ---

# Prepare corpus
dictionary = corpora.Dictionary(df['tokens'])
corpus = [dictionary.doc2bow(text) for text in df['tokens']]

num_topics = 5
lda_model = models.LdaModel(corpus, num_topics=num_topics, id2word=dictionary, passes=10, random_state=42)

# Print topics
for i, topic in lda_model.print_topics(num_words=6):
    print(f"Topic {i}: {topic}")

# Generate descriptive labels for topics
num_top_words = 5
topic_labels = {}
for i in range(num_topics):
    top_words = lda_model.show_topic(i, topn=num_top_words)
    topic_labels[i] = ", ".join([word for word, prob in top_words])

print("\nTopic Labels:")
for idx, label in topic_labels.items():
    print(f"Topic {idx}: {label}")

def assign_lda_category(doc_bow):
    topics = lda_model.get_document_topics(doc_bow)
    if topics:
        dominant_topic = max(topics, key=lambda x: x[1])[0]
        return topic_labels[dominant_topic]
    return 'None'

df['category'] = [assign_lda_category(bow) for bow in corpus]

# --- Manual Dictionary Based Categories (category1) ---
categories = {
    'health': ['health', 'medical', 'patient', 'disease', 'hospital'],
    'finance': ['finance', 'economic', 'stock', 'market', 'price'],
    'sports': ['sport', 'football', 'basketball', 'athlete', 'game'],
    'tech': ['tech', 'computer', 'ai', 'machine', 'learning'],
    'social': ['social', 'media', 'twitter', 'facebook', 'network']
}

def assign_manual_category(text):
    text_tokens = set(text.split())
    for cat, keywords in categories.items():
        if any(keyword in text_tokens for keyword in keywords):
            return cat
    return 'other'

df['category1'] = df['about_dataset'].apply(assign_manual_category)

# --- Fuzzy Matching Based Categories (category2) ---
# This attempts to be more flexible by fuzzy matching tokens with keywords.

def assign_fuzzy_category(text, threshold=80):
    text_tokens = set(text.split())
    best_cat = 'other'
    best_score = 0
    for cat, keywords in categories.items():
        for keyword in keywords:
            for token in text_tokens:
                score = fuzz.ratio(token, keyword)
                if score > best_score and score >= threshold:
                    best_score = score
                    best_cat = cat
    return best_cat

df['category2'] = df['about_dataset'].apply(assign_fuzzy_category)

# Plot LDA categories
plt.figure(figsize=(12, 6))
df['category'].value_counts().plot(kind='bar', color='teal')
plt.title('Dataset Categories by LDA Topics (Descriptive Labels)', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot manual categories
plt.figure(figsize=(12, 6))
df['category1'].value_counts().plot(kind='bar', color='coral')
plt.title('Dataset Categories by Manual Dictionary', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot fuzzy matching categories
plt.figure(figsize=(12, 6))
df['category2'].value_counts().plot(kind='bar', color='purple')
plt.title('Dataset Categories by Fuzzy Matching', fontsize=16)
plt.xlabel('Category', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# Prepare the metadata for embedding and storage in a vector database

```python
# Combine metadata with category2 included
df['combined_text'] = df['header'] + ' ' + df['dataset_name'] + ' ' + df['about_dataset'] + ' ' + df['category2']

# Embed text
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(df['combined_text'].tolist(), show_progress_bar=True)

# Store in FAISS
dimension = embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(embeddings)
faiss.write_index(index, 'dataset_index.faiss')

# Save metadata with embeddings info
df.to_csv('metadata_with_embeddings.csv', index=False)
```

# Retrieval Pipeline with example

```python
import pandas as pd
import numpy as np
```

```python
# Load model (must match the one used for indexing)
model = SentenceTransformer('all-MiniLM-L6-v2')

# Load FAISS index
index = faiss.read_index('dataset_index.faiss')

# Load metadata
df = pd.read_csv('metadata_with_embeddings.csv')

# Function to retrieve top-k most relevant datasets for a user query
def retrieve_datasets(query, top_k=5):
    # Encode query using the same model
    query_embedding = model.encode([query])

    # Search FAISS index
    distances, indices = index.search(np.array(query_embedding), top_k)

    # Extract top-k metadata
    results = df.iloc[indices[0]].copy()
    results['similarity_score'] = 1 - distances[0]  # Optional: convert to similarity
    return results[['dataset_name', 'header','about_dataset','dataset_url', 'category2', 'similarity_score']]

# Example usage
query = "social media usage by teenagers"
top_results = retrieve_datasets(query, top_k=5)
print(top_results)
```

# Augment with EDA Insights

```python
def eda_and_prompt(top_results):
    # Compute description length if not present
    if 'desc_length' not in top_results.columns:
        top_results['desc_length'] = top_results['about_dataset'].apply(lambda x: len(str(x)))

    # Extract top categories and their counts
    top_categories = top_results['category2'].value_counts()
    avg_similarity = top_results['similarity_score'].mean()
    avg_desc_len = top_results['desc_length'].mean()

    # Extract dataset names, headers, descriptions, categories, and URLs (limit to top 5)
    datasets = top_results[['dataset_name', 'header', 'about_dataset', 'category2', 'dataset_url']].head(5)

    # Prepare detailed dataset block for LLM
    dataset_blocks = []
    for i, row in datasets.iterrows():
        block = f"""
📁 **Dataset Name:** {row['dataset_name']}
📌 **Header Summary:** {row['header']}
🗂️ **Category:** {row['category2']}
🔗 **Link:** {row['dataset_url']}
📝 **Description:** {row['about_dataset']}
"""
        dataset_blocks.append(block)

    combined_dataset_info = '\n'.join(dataset_blocks)

    # Compose EDA textual summary
    insights = f"""
📊 **EDA Insights:**
- Most common dataset categories: {', '.join(top_categories.index[:3])}
- Average similarity score across datasets: {avg_similarity:.3f}
- Average description length: {avg_desc_len:.0f} characters

🧠 **Prompt to LLM:**
Below are multiple datasets retrieved from a search. Each includes its name, description, category, and link.

{combined_dataset_info}

Please help with the following:
1. Return the dataset names, url to the dataset and information about the dataset
2. Summarize what each dataset is about in simpler terms.
3. Suggest possible EDA steps or interesting patterns that could be explored based on the description.
4. Recommend project ideas or machine learning use cases aligned with each dataset.
5. If possible, suggest or generate basic plots that could be used for initial EDA using only the description above.

Your responses should reflect the unique nature of each dataset and the category it's part of.
"""

    print(insights)
    return insights
# Usage:
llm_query = eda_and_prompt(top_results)
```
